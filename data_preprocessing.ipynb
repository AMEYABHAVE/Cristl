```python
from google.colab import drive
import pandas as pd
file_location = '/content/drive/MyDrive/accelerometer.csv'
file_location1 = '/content/drive/MyDrive/gyroscope.csv'
accelerometer =pd.read_csv(file_location)
accelerometer.drop(columns=['time','repetition','uuid'], inplace=True)
gyroscope =pd.read_csv(file_location1)
gyroscope.drop(columns=['time','repetition','uuid'], inplace=True)
```

```python
accelerometer.head()
```

```python
gyroscope.head()
```

```python
accchair= accelerometer[accelerometer['activity'] == 'CHAIR']
acccouch= accelerometer[accelerometer['activity'] == 'COUCH']
accsurface= accelerometer[accelerometer['activity'] == 'SURFACE']

gyrchair= gyroscope[gyroscope['activity'] == 'CHAIR']
gyrcouch= gyroscope[gyroscope['activity'] == 'COUCH']
gyrsurface= gyroscope[gyroscope['activity'] == 'SURFACE']
```

```python
import matplotlib.pyplot as plt
plt.title("ACCELEROMETER acc_x")
plt.plot(accchair["acc_x"], label="chair")
plt.plot(acccouch["acc_x"], label="couch")
plt.plot(accsurface["acc_x"], label="surface")
plt.legend()
plt.show()

plt.title("ACCELEROMETER acc_y")
plt.plot(accchair["acc_y"], label="chair")
plt.plot(acccouch["acc_y"], label="couch")
plt.plot(accsurface["acc_y"], label="surface")
plt.legend()
plt.show()

plt.title("ACCELEROMETER acc_z")
plt.plot(accchair["acc_z"], label="chair")
plt.plot(acccouch["acc_z"], label="couch")
plt.plot(accsurface["acc_z"], label="surface")
plt.legend()
plt.show()
```

```python
plt.title("GYROSCOPE gyr_x")
plt.plot(gyrchair["gyr_x"], label="chair")
plt.plot(gyrcouch["gyr_x"], label="couch")
plt.plot(gyrsurface["gyr_x"], label="surface")
plt.legend()
plt.show()

plt.title("GYROSCOPE gyr_y")
plt.plot(gyrchair["gyr_y"], label="chair")
plt.plot(gyrcouch["gyr_y"], label="couch")
plt.plot(gyrsurface["gyr_y"], label="surface")
plt.legend()
plt.show()

plt.title("GYROSCOPE gyr_z")
plt.plot(gyrchair["gyr_z"], label="chair")
plt.plot(gyrcouch["gyr_z"], label="couch")
plt.plot(gyrsurface["gyr_z"], label="surface")
plt.legend()
plt.show()
```

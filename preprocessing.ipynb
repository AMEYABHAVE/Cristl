```python
import pandas as pd
file_location = '/content/drive/MyDrive/cristldataset.csv'
df =pd.read_csv(file_location)
df.head()
```

```python
accelerometer= df[df['name'] == 'accelerometer']
gyroscope= df[df['name'] == 'gyroscope']
name= df[df['name'] == 'name']
df.drop(name.index, inplace=True)

chair= df[df['activity'] == 'CHAIR']
couch= df[df['activity'] == 'COUCH']
surface= df[df['activity'] == 'SURFACE']

accchair= accelerometer[df['activity'] == 'CHAIR']
acccouch= accelerometer[df['activity'] == 'COUCH']
accsurface= accelerometer[df['activity'] == 'SURFACE']

gyrchair= chair[df['name'] == 'gyroscope']
gyrcouch= couch[df['name'] == 'gyroscope']
gyrsurface= surface[df['name'] == 'gyroscope']
```

```python
import matplotlib.pyplot as plt
plt.title("acc_x")
plt.plot(chair["acc_x"], label="chair")
plt.plot(couch["acc_x"], label="couch")
plt.plot(surface["acc_x"], label="surface")
plt.legend()
plt.show()

plt.title("acc_y")
plt.plot(chair["acc_y"], label="chair")
plt.plot(couch["acc_y"], label="couch")
plt.plot(surface["acc_y"], label="surface")
plt.legend()
plt.show()

plt.title("acc_z")
plt.plot(chair["acc_z"], label="chair")
plt.plot(couch["acc_z"], label="couch")
plt.plot(surface["acc_z"], label="surface")
plt.legend()
plt.show()
```

```python
plt.title("ACCELEROMETER acc_x")
plt.plot(accchair["acc_x"], label="chair")
plt.plot(acccouch["acc_x"], label="couch")
plt.plot(accsurface["acc_x"], label="surface")
plt.legend()
plt.show()

plt.title("ACCELEROMETER acc_y")
plt.plot(accchair["acc_y"], label="chair")
plt.plot(acccouch["acc_y"], label="couch")
plt.plot(accsurface["acc_y"], label="surface")
plt.legend()
plt.show()

plt.title("ACCELEROMETER acc_z")
plt.plot(accchair["acc_z"], label="chair")
plt.plot(acccouch["acc_z"], label="couch")
plt.plot(accsurface["acc_z"], label="surface")
plt.legend()
plt.show()
```

```python
plt.title("GYROSCOPE acc_x")
plt.plot(gyrchair["acc_x"], label="chair")
plt.plot(gyrcouch["acc_x"], label="couch")
plt.plot(gyrsurface["acc_x"], label="surface")
plt.legend()
plt.show()

plt.title("GYROSCOPE acc_y")
plt.plot(gyrchair["acc_y"], label="chair")
plt.plot(gyrcouch["acc_y"], label="couch")
plt.plot(gyrsurface["acc_y"], label="surface")
plt.legend()
plt.show()

plt.title("GYROSCOPE acc_z")
plt.plot(gyrchair["acc_z"], label="chair")
plt.plot(gyrcouch["acc_z"], label="couch")
plt.plot(gyrsurface["acc_z"], label="surface")
plt.legend()
plt.show()
```
